```python
import numpy as np
import matplotlib.pyplot as plt

class plotcurves:
    def __init__(self):
        self.x=np.arange(-3,3,0.1)

    
    def plots(self):
        y=np.sin(self.x)
        plt.plot(self.x,y)
        plt.show()
        return
    def plotc(self):
        y=np.cos(self.x)
        plt.plot(self.x,y)
        plt.show()
        return
    
```

```python
p=plotcurves()
```

```python
p.plots()
```

```python
p.plotc()
```
